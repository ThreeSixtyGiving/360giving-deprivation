### Lookup for Post Codes to LAD11 - which is what we will use on our app, the GrantNav data uses LAD17

```python
postcode_lad11_df = pd.read_csv('../data/geo/pcd11_par11_wd11_lad11_ew_lu.csv',encoding='iso-8859-1')
```

```python
postcode_lad11_df.head()
```

```python
oa_to_ward_2017 = pd.read_csv('../data/geo/Output_Area_to_Ward_to_Local_Authority_District_December_2017_Lookup_in_England_and_Wales.csv',encoding='iso-8859-1')
```

```python
oa_to_ward_2017.head()
```

```python

lad17_df = pd.DataFrame.pivot_table(oa_to_ward_2017, columns=['LAD17CD','LAD17NM'],
                      aggfunc='count', fill_value=None, margins=False,
                      dropna=True, margins_name='All')['OA11CD'].reset_index()
```

```python
lad17_df.head()
```

```python
lad11_df = pd.DataFrame.pivot_table(postcode_lad11_df, columns=['lad11cd','lad11nm'],
                      aggfunc='count', fill_value=None, margins=False,
                      dropna=True, margins_name='All')['pcd8'].reset_index()
```

```python
lad11_df.head()
```

```python

lad17_to_lad11_lookup = pd.merge(lad11_df, lad17_df, left_on='lad11nm', right_on='LAD17NM',sort=True)
```

```python
df = pd.merge(df, lad17_to_lad11_lookup, left_on='Recipient District Geographic Code', right_on='LAD17CD',sort=True)
```

```python
df[df['Recipient District'] == 'St. Albans District']
```

```python
df.head()
```

```python
aggregagations = {'Identifier': 'count',      
                  'Amount Awarded': 'sum'}

themes_pivot_df = pd.pivot_table(df, index=['Funding Org:Name', 'Award Year', 'lad11cd','Recipient Ward Geographic Code','Recipient Region','Recipient District'], aggfunc = aggregagations)
#themes_pivot_df = pd.pivot_table(df, index=['Funding Org:Name', 'Award Year'], aggfunc = aggregagations)
themes_pivot_df.reset_index(inplace=True)
themes_pivot_df.rename(columns={'lad11cd': 'Recipient District Geographic Code'},inplace=True)
```

```python
themes_pivot_df.head()
```

```python
out = '[' + themes_pivot_df.to_json(orient='records')[1:-1] + ']'
with open('../data/deprivation/grants_eng_LAD_ward_donor_year.json', 'w') as f:
    f.write(out)
```

## Aggregated deprivation by LAD

```python
aggregagations = {'Identifier': 'count',      
                  'Amount Awarded': 'sum'}

lad_award_pivot_df = pd.pivot_table(themes_pivot_df, index=['Recipient District Geographic Code'], aggfunc = aggregagations)
#themes_pivot_df = pd.pivot_table(df, index=['Funding Org:Name', 'Award Year'], aggfunc = aggregagations)
lad_award_pivot_df.reset_index(inplace=True)
```

```python
lad_award_pivot_df.head()
```

## Merge with LAD deprivation data

```python
lad_vars = [
    {'sheet': 'IMD', 'col': 'IMD - Average score'},
    {'sheet': 'Income', 'col': 'Income - Average score'},
    {'sheet': 'Employment', 'col': 'Employment - Average score'},
    {'sheet': 'Education', 'col': 'Education, Skills and Training - Average score'},
    {'sheet': 'Health', 'col': 'Health Deprivation and Disability - Average score'},
    {'sheet': 'Crime', 'col': 'Crime - Average score'},
    {'sheet': 'Barriers', 'col': 'Barriers to Housing and Services - Average score'},
    {'sheet': 'Living', 'col': 'Living Environment - Average score'}
]
```

```python
lad_df = pd.DataFrame()
```

```python
for i, v in enumerate(lad_vars):
    #print(v['sheet'],i)
    df = pd.read_excel('../data/deprivation/File_10_ID2015_Local_Authority_District_Summaries.xlsx', sheet_name=v['sheet'])
    df.rename(index=str, columns={'Local Authority District code (2013)' : 'LAD2013_CD', \
        'Local Authority District name (2013)' : 'LAD2013_NM', \
        v['col'] : v['sheet']+'_avg_score', \
        }, inplace=True)
    if i == 0:
        lad_df = lad_df.append(df[['LAD2013_CD','LAD2013_NM',v['sheet']+'_avg_score']])
    else:
        lad_df = pd.merge(lad_df, df[['LAD2013_CD',v['sheet']+'_avg_score']], on='LAD2013_CD',sort=True)
    
#dp_lad_df = pd.read_excel('../data/deprivation/File_10_ID2015_Local_Authority_District_Summaries.xlsx', sheet_name='ID2015 Domains', index_col=[0])

display(lad_df.head())
```

```python
combined_gpd = pd.DataFrame()
combined_gpd = pd.merge(lad_award_pivot_df, lad_df, \
                  left_on='Recipient District Geographic Code', right_on='LAD2013_CD',sort=True)
```

```python
combined_gpd.head()
```

```python
out = '[' + combined_gpd.to_json(orient='records')[1:-1] + ']'
with open('../data/deprivation/grants_by_lad_deprivation.json', 'w') as f:
    f.write(out)
```
